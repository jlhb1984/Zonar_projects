```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
esp840=pd.read_csv(r'C:\Users\Jose.Hurtado\Documents\projects\data\zona_logistica\esp840.csv',sep=',')
esp840.info()
```

```python
aux=[]
aux_miles=[]
x_axis=[]
esp840_odometer=pd.DataFrame()
cont=0

for i in range (0, esp840.shape[0]):
    l=len(esp840.iloc[i,2])
    #print('Len=',l)
    if l<180:
        aux.append(int(esp840.iloc[i,2][68:76],16))
        aux_miles.append((int(esp840.iloc[i,2][68:76],16))*0.621371)
        cont=cont+1 

esp840_odometer['Odometer in']=aux
esp840_odometer['Odometer en miles']=aux_miles

for i in range(0,cont):
    x_axis.append(i)

esp840_odometer
plt.plot(x_axis,aux)
```

```python
esp840=pd.read_csv(r'C:\Users\Jose.Hurtado\Documents\projects\data\zona_logistica\esp840_1620_02_2026.csv',sep=',')
esp840.info()
```

```python
aux=[]
aux_miles=[]
x_axis=[]
date=[]
esp840_odometer=pd.DataFrame()
cont=0

for i in range (0, esp840.shape[0]):
    l=len(esp840.iloc[i,2])
    #print('Len=',l)
    if l<180:
        aux.append(int(esp840.iloc[i,2][68:76],16))
        aux_miles.append((int(esp840.iloc[i,2][68:76],16))*0.000621371)
        date.append(esp840.iloc[i,0])
        cont=cont+1 

for i in range(0,cont):
    x_axis.append(i)

esp840_odometer['Odometer in meters']=aux
esp840_odometer['Odometer en miles']=aux_miles
esp840_odometer['Date']=date
esp840_odometer['x']=x_axis

esp840_odometer.to_csv(r'C:\Users\Jose.Hurtado\Documents\projects\data\zona_logistica\esp840_1620_02_2026_data.csv')

esp840_odometer
plt.plot(x_axis,aux)
```
